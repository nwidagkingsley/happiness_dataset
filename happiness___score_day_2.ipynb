OUTLIER DETECTION

```python
Q1 = df['Happiness Score'].quantile(0.25)
Q3 = df['Happiness Score'].quantile(0.75)
IQR =df['Happiness Score'].quantile(0.5)

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Happiness Score'] < lower) | (df['Happiness Score'] > upper)]
outliers
```

```python
upper = Q3 + 3.0 * IQR
print(upper)
```

```python
lower = Q1 - 1.5 * IQR
print(lower)
```
